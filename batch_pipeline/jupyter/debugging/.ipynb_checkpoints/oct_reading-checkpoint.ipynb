```python
import zipfile
import os

def unzip_oct_file(zip_path, extract_path):
    """
    Unzips an .oct file which is in a ZIP format.

    :param zip_path: Path to the .oct ZIP file
    :param extract_path: Directory where contents will be extracted
    """
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
            print(f"Files extracted to: {extract_path}")
    except zipfile.BadZipFile:
        print("Error: The file is not a zip file or it is corrupted.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the path to your .oct file and the directory to extract its contents
zip_path = '/project/dkarig/ecocoat/data/oct/12_9_24 Flow Cells/AYLFD86_24_12_10_18_47_03.oct'
extract_path = './test'

# Create the directory if it doesn't exist
if not os.path.exists(extract_path):
    os.makedirs(extract_path)

unzip_oct_file(zip_path, extract_path)
```
