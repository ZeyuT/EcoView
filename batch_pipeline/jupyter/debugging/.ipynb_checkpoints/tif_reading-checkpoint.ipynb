## Read in b-scan sequences

```python
def readImageSequence(filelist,dtype):
    filelist=glob.glob(path+'/*.tif')
    filelist.sort()
    stack = []
    for file in filelist:
        stack.append(imread(file).astype(dtype))
    stack = np.transpose(stack, (0,2,1))
    return np.array(stack)
```

```python
def read_tif_stack(args):
    i, file_path = args
    page = TiffFile(file_path).pages[i].asarray()
    # In rare cases, the last one or two pages are empty
    if page is not None:
        return i, page[:,:,:]
def read_tiff(file_path):    
    # Read pages in tiff file using multiprocessing
    pages = TiffFile(file_path).pages
    n_threads = min(16,mp.cpu_count()-2)
    print(f'# threads: {n_threads}')

    pool = mp.Pool(n_threads)
    results = pool.map(read_tif_stack, [[i, file_path] for i in range(len(pages))])
    pool.close()
    pool.join()
    results.sort(key=lambda x:x[0])
    vol = []
    for _, frame in results:
        vol.append(frame)
    return np.array(vol)
```

```python
imwrite('./temp/control.tif', cropped.astype('uint8'))
```

```python
binary.shape
```

```python
def get_column_median(col,max_h,min_h):
    nonzero_pos = sorted(np.where(col[min_h:max_h]))[0]
    return nonzero_pos[len(nonzero_pos)//2]+min_h 
```

```python
raw_plane = np.zeros((binary.shape[0],binary.shape[2]))
```

```python
for x in range(binary.shape[0]):
    for y in range(binary.shape[2]):
        raw_plane[x,y] = get_column_median(col=binary[x,:,y],
                                           max_h=250,
                                           min_h=150)

plane = ndimage.median_filter(input=raw_plane, 
                             size=21,
                             mode='reflect')
```

```python
offset = (plane - np.min(plane)).astype('uint8')
imwrite('./temp/offset.tiff', offset.astype('uint8'))
```

```python
plt.imshow(binary[200]*255,cmap='gray', vmin=0, vmax=255)
plt.plot(plane[200])
plt.axhline(100,c='g')
plt.axhline(270,c='g')
```

```python
def correction(vol, offset):
    output = np.zeros_like(vol)
    for x in range(offset.shape[0]):
        for y in range(offset.shape[1]):
            cur_offset = offset[x,y]
            output[x,:output.shape[1]-cur_offset,y] = vol[x,cur_offset:,y]
    return output
```

```python
corrected_controled = correction(cropped, offset)
```
