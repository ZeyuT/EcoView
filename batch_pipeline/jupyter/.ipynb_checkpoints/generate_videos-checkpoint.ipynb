### Make videos from shear images

```python
fps = 24
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

image_path = '/project/ahoover/mhealth/zeyut/biofilm/2024-12/results_12_9_24 Flow Cells/Shear/height_maps/'
tmp_output_path = './results'
output_path = '/project/ahoover/mhealth/zeyut/biofilm/2024-12/results_12_9_24 Flow Cells/videos_top_view'
```

```python
%autoreload 1
%aimport video_helpers
```

```python
all_files = glob.glob(image_path + '*')
marks = ['Before shear','During shear','After shear', '24 hrs post shear']
position = [500, 30]
thickness = 2
color = (144, 238, 144)
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
line_type = cv2.LINE_AA

for community in ['AYLFD86S', 'BP1FD86S', 'DBSFD85S', 'FCCFD81S', 'HALFD86S', 'KARFD83S']:
# for community in ['KARFD83S']:
    frame_list = []
    # Read in the size of the video made of multi-days data
    cap = cv2.VideoCapture(os.path.join(output_path, f'{community}.mp4'))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cap.release()
    text_list = []
    for i, tag in enumerate(['Before Shear', 'During Shear', 'After Shear', 'Final']):
        file = os.path.join(image_path, f'12_9_24 Flow Cells_{community}_{tag}.jpg')
        if not os.path.exists(file):
            print(f'Cannot find {tag} image for {community}')
            continue
        # Read and resize frames
        frame = cv2.imread(file, cv2.IMREAD_UNCHANGED)
        frame = cv2.resize(frame, (width, height), interpolation=cv2.INTER_AREA)
        text_list.append(marks[i])
        frame_list.append(frame) 
    frame_list = np.array(frame_list)
    # stabilize frames
    # frame_list = video_helpers.stabilize_frames_StackReg(frame_list, 20)

    # Stretch frame values
    if frame_list.max()>0:
        max_ = np.percentile(frame_list[frame_list!=0], 99.9)
        frame_list = np.clip(frame_list, 0, max_)    
        frame_list = (frame_list - frame_list.min())/(max_-frame_list.min()) * 255.
    frame_list = frame_list.astype('uint8')
    
    proc_frame_list = []
    for i in range(len(frame_list)):
        (text_width, text_height), baseline = cv2.getTextSize(text_list[i], font_face, font_scale, thickness)
        proc_frame = cv2.cvtColor(frame_list[i], cv2.COLOR_GRAY2BGR)
        proc_frame = cv2.putText(proc_frame, text_list[i], 
                            (position[0]-text_width, position[1]), 
                            font_face, font_scale, 
                            color, thickness, 
                            line_type)
        proc_frame_list += [proc_frame]*8 # Display each key frame for 2 sec
    proc_frame_list = np.array(proc_frame_list)
    # Add marks        
    video_helpers.create_video(proc_frame_list, 
                os.path.join(tmp_output_path, f'shear_{community}.mp4'),
                fourcc=fourcc, 
                num_transitions=6, 
                fps=24)

```

### Add shear video to the end of multi-day video and overwrite 

```python
# Add shear video to the end of multi-day video and overwrite 
for community in ['AYLFD86S', 'BP1FD86S', 'DBSFD85S', 'FCCFD81S', 'HALFD86S', 'KARFD83S']:
    # Copy the multi-day video to a temporary directory
    shutil.copy(os.path.join(output_path, f'{community}.mp4'), os.path.join(tmp_output_path, f'{community}.mp4'))
    cap1 = cv2.VideoCapture(os.path.join(tmp_output_path, f'{community}.mp4')) # Multi-day videos, grayscale
    cap2 = cv2.VideoCapture(os.path.join(tmp_output_path, f'shear_{community}.mp4')) # Multi-day videos, Color
    width = int(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Overwrite the original video
    out = cv2.VideoWriter(os.path.join(output_path, f'{community}.mp4'), fourcc, fps, (width, height),isColor=True)
    while True:
        ret, frame = cap1.read()
        if not ret:
            break
        out.write(frame)
    while True:
        ret, frame = cap2.read()
        if not ret:
            break
        out.write(frame)
    out.release()
    cap1.release()
    cap2.release()
    # Remove the copied video in temporary directory
    os.remove(os.path.join(tmp_output_path, f'{community}.mp4'))
    # Remove the shear video in temporary directory
    os.remove(os.path.join(tmp_output_path, f'shear_{community}.mp4'))
```
