### Measure sub-areas on shear images

```python
def split_into_patches(volume, grid_size=(4, 4), overlap=0.75):
    # Assuming volume is a numpy array of shape [Z, X, Y]
    patches = []
    patch_size = [volume.shape[1]//grid_size[0], volume.shape[2]//grid_size[1]]
    x_step = int(volume.shape[1] // grid_size[0] * (1-overlap))
    y_step = int(volume.shape[2] // grid_size[1] * (1-overlap))
    locations = []
    for x_start in range(0, volume.shape[1]-patch_size[0]+1, x_step):
        for y_start in range(0, volume.shape[2]-patch_size[1]+1, y_step):
            patch = volume[:, x_start:x_start + patch_size[0], y_start:y_start + patch_size[1]]
            patches.append(patch)
            locations.append([x_start, y_start])
    return np.array(patches), locations
```

```python
root_path = '/project/ahoover/mhealth/zeyut/biofilm/2024-12/results_12_9_24 Flow Cells/Shear/height_maps/'
file_list = glob.glob(os.path.join(root_path, '*Shear*')) + glob.glob(os.path.join(root_path, '*Final*'))
file_list.sort()
```

```python
grid_size = (4, 4)
```

```python
excel_writer = pd.ExcelWriter(os.path.join('./','results_grid_measures.xlsx'),mode='w') 

for file_path in file_list:
    community = file_path.split('Flow Cells_')[1].split('.jpg')[0]
    print(community)
    height_map = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    x_size = height_map.shape[1] 
    y_size = height_map.shape[0] 
    x_step = x_size // grid_size[1]
    y_step = y_size // grid_size[0]                 
    patch_size = [x_size//grid_size[0], y_size//grid_size[1]]
            
    locations = []
    df_res = pd.DataFrame()
    for y_start in range(0, y_size-1, y_step):
        for x_start in range(0, x_size-1, x_step):
            patch = height_map[y_start:y_start + patch_size[1], x_start:x_start + patch_size[0]]
            cur_res = measure_morphology(patch)
            cur_res['row_idx'] = y_start//y_step
            cur_res['col_idx'] = x_start//x_step
            df_res = df_res.append(cur_res, ignore_index=True)
            df_res['row_idx'] = df_res['row_idx'].astype(int)
            df_res['col_idx'] = df_res['col_idx'].astype(int)
    df_res.to_excel(excel_writer, sheet_name=community, index=False)
excel_writer.close()
```
